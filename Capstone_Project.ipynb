# To Clean Data

* Removal of unrequired features

```python
# Removal of features that should not be used since they do not help predict creditworthiness for a new application.

df_copy = df_copy.drop("id", axis=1) # Exclude id since its an identifier alone and not predictive
df_copy = df_copy.drop("address_state", axis=1) # Exclude id since its an identifier alone and not predictive
df_copy = df_copy.drop("emp_title", axis=1) # Exclude since this is an identifier alone
df_copy = df_copy.drop("grade", axis=1) # Exclude as this is not required
df_copy = df_copy.drop("issue_date", axis=1) # Exclude since this has no effect on the prediction
df_copy = df_copy.drop("last_credit_pull_date", axis=1) # Exclude happens after loan is issued
df_copy = df_copy.drop("last_payment_date", axis=1) # Exclude as this is determined after repayment of loan
df_copy = df_copy.drop("next_payment_date", axis=1) # Exclude since its a future information
df_copy = df_copy.drop("total_payment", axis=1) # Exclude since its just an outcome after a loan
df_copy = df_copy.drop("member_id", axis=1) # Exclude since its an identifier alone and not predictive
df_copy = df_copy.drop("sub_grade", axis=1) # Exclude as this is not required
df_copy = df_copy.drop("application_type", axis=1) # Exclude since we have just an application type
```
