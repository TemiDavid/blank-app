### Check for Outliers

```python
df_copy.describe()
```

Using the IQR (Inter-Quartile Range) method, it is noticed that there are outliers in the annual_income, installment, int_rate, loan)amount and total_acc columns.

The steps below is to automatically detect and treat outliers in all numeric columns in the cleaned dataset using the IQR (Interquartile Range) method,

```python
# Automatically Detect Numeric Columns
numeric_cols = df_copy.select_dtypes(include=['int64','float64']).columns

print("Numeric Columns:")
print(numeric_cols)
```

```python
# Detect Outliers in All Numeric Columns
outlier_summary = {}

for col in numeric_cols:

    Q1 = df_copy[col].quantile(0.25)
    Q3 = df_copy[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df_copy[(df_copy[col] < lower) | (df_copy[col] > upper)]

    outlier_summary[col] = len(outliers)

print("Outlier Count Per Column")
print(outlier_summary)
```

```python
# Automatically Treat Outliers (Capping Method)
for col in numeric_cols:

    Q1 = df_copy[col].quantile(0.25)
    Q3 = df_copy[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_copy[col] = df_copy[col].clip(lower, upper)

print("Outliers treated successfully.")
```

```python
# To verify That Outliers Are Removed
check_outliers = {}

for col in numeric_cols:

    Q1 = df_copy[col].quantile(0.25)
    Q3 = df_copy[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df_copy[(df_copy[col] < lower) | (df_copy[col] > upper)]

    check_outliers[col] = len(outliers)

print("Remaining Outliers After Treatment")
print(check_outliers)
```
